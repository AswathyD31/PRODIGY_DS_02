# **DATA CLEANING**

```python
print("Missing values in each column:\n")
print(df.isnull().sum())
```

```python
# Fill 'Age' with median as it's numerical
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
# Fill 'Embarked' with mode as it's categorical
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# Drop 'Cabin' due to too many missing values
df.drop(columns='Cabin', inplace=True)
```

```python
df
```

### Note:

* Missing values in Age are filled with median.

* Missing Embarked values are filled with the mode.

* Cabin column is dropped due to excessive missing data.

# EXPLORATORY DATA ANALYSIS (EDA)
________

________
## 1. TARGET VARIABLE DISTRIBUTION (Survived)
________

```python
sns.countplot(x='Survived', data=df, palette='Set1')
plt.title('Survival Count')
plt.show()

# Survival rate
print(df['Survived'].value_counts(normalize=True))
```

**Insight:**

* Overall survival rate was around 38% to 40%, meaning the majority of passengers did not survive.

* The dataset is moderately imbalanced.

________
## 2. SURVIVAL RATE BY SEX
________

```python
sns.countplot(x='Survived', hue='Sex', data=df, palette='viridis')
plt.title('Survival Count by Gender')
plt.show()
```

**Insight:**

* Females had a significantly higher survival rate compared to males.

* Majority of the deceased passengers were male.

**Conclusion:** Gender played a crucial role in survival; women were prioritized during evacuation.

________
## 3. SURVIVAL RATE BY PASSENGER CLASS
________

```python
sns.countplot(x='Survived', hue='Pclass', data=df, palette='muted')
plt.title('Survival Count by Passenger Class')
plt.show()
```

**Insight:**

* 1st class passengers had the highest survival rate.

* 3rd class passengers had the lowest survival rate.

**Conclusion:**  Socio-economic status (as reflected by class) influenced survival chances.

________
## 4. AGE DISTRIBUTION
________

```python
sns.histplot(df['Age'], kde=True, color='green')
plt.title('Age Distribution')
plt.show()
```

**Insight:**

* Most passengers were between 20 to 40 years old.

* Some children and elderly passengers were present, though fewer in number.

**Conclusion:** The Titanic carried predominantly young to middle-aged adults.



________
## 5. SURVIVAL BY EMBARKATION POINT
________

```python
sns.countplot(x='Survived', hue='Embarked', data=df, palette='pastel')
plt.title('Survival Count by Embarkation Point')
plt.show()
```

**Insight:**

* Passengers who embarked at Cherbourg (C) had a higher survival rate.

* Southampton (S) passengers had the lowest survival rate.

**Conclusion:**  Embarkation point could be linked to socio-economic status or cabin location.

________
# 6.  FARE DISTRIBUTION BY CLASS
________

```python
sns.violinplot(x='Pclass', y='Fare', data=df, palette='Spectral')
plt.title('Fare Density by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

**Insight:**

* 1st class passengers paid significantly higher fares, with a wider range of ticket prices.

* 3rd class fares were the lowest with little variation.

**Conclusion:** Ticket fare positively correlates with passenger class.

________
# 7. FARE DISTRIBUTION BY SURVIVAL
________

```python
sns.boxplot(x='Survived', y='Fare', data=df, palette=['salmon', 'green'])
plt.title('Fare Paid by Survival')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()
```

**Insight:**

* Survivors tended to have paid higher fares on average compared to those who did not survive.

* Some lower-fare passengers survived, but it was less common.

**Conclusion:**  Paying more (likely for better cabins) increased chances of survival.

________
# 8. CORRELATION HEATMAP
________

```python
corr = df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

**Insight:**

* Fare and Pclass are negatively correlated (higher class → higher fare).

* Survival shows weak positive correlations with Fare and weak negative correlations with Pclass.

**Conclusion:**  Higher fare and class increased survival likelihood, though no single numerical factor was dominant.

________
# 9. PAIRPLOT FOR OVERALL PATTERNS
________

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']], hue='Survived', palette='husl')
plt.show()
```

**Insight:**

* Clear separation between survival categories across Age, Fare, and Pclass.

* Surviving passengers cluster at lower Pclass, higher Fare, and wider Age ranges.

**Conclusion:**  Multiple variables together better explain survival compared to any one feature alone.



________
# 10. MULTIVARIATE RELATIONSHIPS: AGE, FARE, CLASS, SURVIVAL
________

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', style='Pclass', palette='cool')
plt.title('Age vs Fare by Survival & Class')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

**Insight:**

* Survivors cluster in younger age ranges with higher fares, especially in 1st and 2nd class.

* Many deceased passengers paid low fares and were from 3rd class.

**Conclusion:**  Survival likelihood increases for younger, wealthier passengers in higher classes.
